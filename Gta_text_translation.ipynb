```python
import pandas as pd
import numpy as np
import re
import time
```

## Importing Scraped google reviews

```python
gta = pd.read_csv('Gta_rating.csv')
```

```python
gta.info()
```

```python
comments = gta['comments'].astype(str)
```

```python
gta.head()
```

## Preprocessing the scraped reviews

```python
for i,comment in enumerate(comments):
    
    comment = str.lower(comment)
    
    x = re.search(r'(.*?)\\n', comment) #Removing Name of the reviewer
    print("Name:", x.group(1))
    comment = comment.replace(x.group(1), '')
    x = re.search(r'\\n(.*?)\\n', comment) #Removing time of the posted review
    print("When:", x.group(1))
    comment = comment.replace(x.group(1),'')
    
    

    
    try:
        x = re.search(r'(?s:.*)\\n(.*?) people found this helpful', comment) #Removing irrelevant text
        pattern = '\\n'+x.group(1) + ' people found this helpful.'
        comment = comment.replace(pattern,'')
    except:
        pass
    
    try:
        pattern = '\\n1 person found this helpful.' #Removing irrelevant text
        comment = comment.replace(pattern,'')
        pattern = '1 person found this helpful.' #Removing irrelevant text
        comment = comment.replace(pattern,'')
    except:
        pass
    
    try:
        pattern = 'was this review helpful to you?' #Removing irrelevant text
        comment = re.sub(pattern,'',comment)        
    except:
        pass
    
    pattern = '\\nhelpful' #Removing irrelevant text
    comment = comment.replace(pattern,'')
    pattern = '\\nunhelpful' #Removing irrelevant text
    comment = comment.replace(pattern,'')
    pattern='\\n'
    comment = comment.replace(pattern,'')
    comment = str.lower(comment)
    
    comments[i] = comment
    print("Parsed:",comment)
    print("-----------------------------------------------------------------------")
```

## Detect language and translate it to English

### Importing libraries for detection and translation

```python
from langdetect import detect
import translators as ts
import translators.server as tss
```

```python
for i,comment in enumerate(comments):
    try:
        result_lang = detect(comment) #Detects language 
        res=''
        if result_lang != 'en': #if the detected language is not english print it.
            print("Text:",comment)
            print("Language code:",result_lang,", Index", i)

            res = tss.google(comment, result_lang, 'en') #using ggogle translator
            
            print("Translated text:\n",res) #print the translated text
            comments[i] = res
            
            if res=='':
                print("Not translated")
                
    except Exception as e: #exception while tranlating 
        print("Not translated")
        print(e)
    if result_lang != 'en':    
        print("-----------------------------------------------------------------")
    time.sleep(1) # Adding sleep because too many request blocks the requests and gives an exception
```
