## Detect language and translate it to English

### Importing libraries for detection and translation

```python
comments = amazon['comments'].astype(str)
```

```python
from langdetect import detect
import translators as ts
import translators.server as tss
```

### Detecting and Translating

```python
for i,comment in enumerate(comments):
    try:
        result_lang = detect(comment) #Detects language 
        res=''
        if result_lang != 'en': #if the detected language is not english print it.
            print("Text:",comment)
            print("Language code:",result_lang,", Index", i) #print index 

            res = tss.google(comment, result_lang, 'en') #using google translator
            
            print("Translated text:\n",res) #print the translated text
            comments[i] = res
            
            if res=='':
                print("Not translated")
                
    except Exception as e: #exception while tranlating print it 
        print("Not translated")
        print(e)
    if result_lang != 'en':    
        print("-----------------------------------------------------------------")
    time.sleep(1) # Adding sleep because too many request blocks the requests and gives an exception
```

```python
amazon["Comments_translated"] = comments
```

```python
amazon.to_csv('amazon_translated_text.csv', index=False ) #saving it to compare
```

## Converting Emoji to text

```python
import emoji
```

```python
print(emoji.demojize("Good ğŸ‘ğŸ½ğŸ‘ğŸ½")) #example
```

```python
comments = amazon['Comments_translated'].astype(str)
helpful = amazon['helpful']
for i,comment in enumerate(comments):
    comment = emoji.demojize(comment) #Converting emojis to text
    if helpful[i]=='One': #Helpfulness is One then convert it to 1
        helpful[i] = '1' 
    comment = re.sub(r'<br>','', comment)
    comments[i] = comment
```

```python
amazon_translated_text = amazon.copy()
amazon_translated_text['comments'] = comments
amazon_translated_text['helpful'] = helpful
```

```python
del amazon_translated_text['Comments_translated']
```
